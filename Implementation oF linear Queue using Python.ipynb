```python
#Implementation oF linear Queue
class LinearQueue:
    def __init__(self,size):
        self.q = [None]*size
        self.front = -1
        self.rear = -1
        self.size = size
    def isempty(self):
        #Time Complexity: O(1)
        return self.front==-1  and self.rear==-1
    def isfull(self):
       #Time Complexity: O(1)
        return self.size==self.rear+1
    def enqueue(self,ele):
        #Time Complexity: O(1)
        if self.isfull():
            raise Exception("Queue is full")
        elif self.isempty():
            self.front,self.rear = 0,0
            self.q[self.rear] = ele
        else:
            self.rear += 1
            self.q[self.rear] = ele
    def dequeue(self):
        #Time Complexity: O(1)
        if self.isempty():
            raise Exception("Queue is empty")
        elif self.front == self.rear:
            t = self.q[self.front]
            self.front = self.rear = -1
            return t
        else:
            t = self.q[self.front]
            self.front += 1
            return t
    def peek(self):
        #Time Complexity: O(1)
        return self.q[self.front]
    def len(self)  :
        #Time Complexity: O(1)
        return self.rear+1 - self.front
    def print(self):
        #Time Complexity: O(1)
        for i in range(self.front,self.rear+1):
            print(self.q[i],end = " ")
#Space Complexity: O(n)
```

```python
a=LinearQueue(5)
a.enqueue(1)
a.enqueue(2)
a.enqueue(3)
a.enqueue(4)
a.enqueue(5)
a.print()
```
